```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, export_graphviz, plot_tree
import pydotplus
from IPython.display import Image
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import tree

dp = pd.read_csv('heart-disease-dataset2.csv')

print('Dataset Length: ', len(dp))
print('Dataset Shape: ', dp.shape)

print(dp.head)

X = pd.get_dummies(dp.drop('result', axis=1), drop_first = True)
Y = dp['result']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=44)
model = DecisionTreeClassifier(criterion='entropy')
ds_train = model.fit(X_train, Y_train)
predictions =  model.predict(X_test)
print(predictions)
X_train.columns

print('Accuracy is')
score = ds_train.score(X_test, Y_test)
print(score)

dot_data = export_graphviz(ds_train, out_file=None, feature_names=X_train.columns, rounded= True, filled = True)
graph = pydotplus.graph_from_dot_data(dot_data)
print(graph)
Image(graph.create_png())

                           
```
